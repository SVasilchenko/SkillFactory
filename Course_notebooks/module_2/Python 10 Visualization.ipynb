```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('tips.xls')
df.sample(5)
df.info()
```

```python
df['total_bill'].sort_values(ascending=False).iloc[0]
```

```python
%matplotlib inline
```

```python
df.plot()
```

```python
df['total_bill'].plot(kind='hist')
```

```python
df[['total_bill','tip']].plot(kind='hist', grid=True, subplots=True, title=['  ',''])
```

```python
df['day'].value_counts().plot(kind = 'bar', 
                              grid = True, 
                              colormap = 'coolwarm', 
                              title = '   ')
```

```python
df.groupby(by=['day','sex'])['total_bill'].mean().plot(kind='bar',
                                          grid=True,
                                          title = '  ')
```

```python
df.plot(x = 'total_bill', 
        y = 'tip', 
        kind = 'scatter', 
        grid = True, 
        title = '   Vs  ')
```

```python
df.pivot_table(index='day',
              values=['total_bill', 'tip'],
              aggfunc = 'sum').plot(kind='bar')
```

```python
df.pivot_table(values = 'tip', index = ['sex', 'smoker'], aggfunc = 'count').plot(kind = 'barh')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.scatter(x = df['total_bill'], y = df['tip'], marker = 'o')
axes.set_title('   Vs C ')
axes.set_xlabel(' ')
axes.set_ylabel(' ')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.hist(df['total_bill'], bins = 20, color = 'green')
axes.set_title('  ')
axes.set_ylabel(' ')
axes.set_xlabel(' , USD')
```
