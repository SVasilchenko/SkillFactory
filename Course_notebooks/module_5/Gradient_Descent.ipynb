```python
def f(x, y):
    return (1 - x) ** 2 + 100 * (y - x ** 2) ** 2
def grad(x, y):
    dx = 2 * x - 2 + 100 * (-4 * y * x + 4 * x ** 3)
    dy = 100 * (2 * y - 2 * x ** 2)
    return (dx, dy)
```

```python
#функция для задания 7.3
def f(x, y, z):
    return x**3 - 2*x**2 +y**2 + z**2 -2*x*y + x*z - y*z + 3*z 
def grad(x, y):
    dx = 6*x + 18*y -18.2 #частные производные df/dx и df/dy
    dy = 18*x + 70*y -62.6
    return (dx, dy)
```

```python
def dist(x1, x2):
    return (x1[0] - x2[0]) ** 2 + (x1[1] - x2[1]) ** 2
```

```python
#функция для задания 7.2
def f(x, y):
    return (2.1 - x - y)**2+ (2.9 - 3*y - x)**2 + (4.1 - x - 5*y)**2
def grad(x, y):
    dx = 6*x + 18*y -18.2 #частные производные df/dx и df/dy
    dy = dy = 18*x + 70*y -62.6
    return (dx, dy)
```

```python
x0 = (1, 1)
gamma = 0.002
alpha = 0.9
x_cur = x0
x_old = x0
vals = []
coords = []
i = 0
while True:
    x_new = (x_cur[0] - gamma * grad(*x_cur)[0]+alpha*(x_cur[0] - x_old[0]),
            x_cur[1] - gamma * grad(*x_cur)[1]+alpha*(x_cur[1] - x_old[1]))
    if dist(x_new, x_cur) < 1e-9:
        break
    x_old = x_cur
    x_cur = x_new
    vals.append(f(*x_cur))
    coords.append(x_cur)
    i += 1
    print(f"iter={i}; x=({x_cur[0]:.2f}, {x_cur[1]:.2f});"
          f" f(x)={f(*x_cur):.2f}; grad f(x)=({grad(*x_cur)[0]:.2f}, {grad(*x_cur)[1]:.2f})")
```
