### Аналогично задаче 2.11.6 постройте OLS регрессию Chance of Admission на выбранные в задаче 2.11.5 признаки, однако теперь сперва стандартизируйте эти признаки.

```python
Data = df_no_res[['toefl','cgpa','target']]
toefl_array = np.array(Data)[:,0]
```

```python
def norm_data(df):
    buffer = pd.DataFrame()
    dict_out = {'data_norm':'', 'mean':{}, 'norm':{}}
    mean_array = np.empty(len(df.columns))
    norm_array = np.empty(len(df.columns))
    
    for column in df:
        dict_out['mean'][column] = df[column].mean()
        dict_out['norm'][column] = np.linalg.norm(df[column] - dict_out['mean'][column])
        buffer[column] = (df[column] - dict_out['mean'][column])/dict_out['norm'][column]
    dict_out['data_norm'] = buffer
    
    return dict_out
    

dict_data_norm = norm_data(Data)
new = np.array([107, 9.1])
new_norm = (new - np.fromiter(dict_data_norm['mean'].values(), dtype=float)[:-1])/np.fromiter(dict_data_norm['norm'].values(), dtype=float)[:-1]
```

```python
y_norm = dict_data_norm['data_norm']['target']
A_norm = np.array(dict_data_norm['data_norm'].drop('target', axis=1, inplace=False))
dict_data_norm['mean']['toefl']
```
